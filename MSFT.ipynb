```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipeline = Pipeline([
    ('std_scaler', StandardScaler())
])
```

```python
msft_labels = train_set['1. open'].copy()
```

```python
msft_labels
```

```python
msft = msft.drop(['1. open'], axis = 1)
```

```python
msft
```

```python
msft_prepared = pipeline.fit_transform(msft)
```

```python
msft_prepared
```

```python
msft_labels
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()
```

```python
lin_reg.fit(msft_prepared, msft_labels)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
msft_predictions = lin_reg.predict(msft_prepared)
```

```python
lin_rmse = mean_squared_error(msft_labels, msft_predictions)
lin_rmse = np.sqrt(lin_rmse)
```

```python
lin_rmse
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
tree_reg = DecisionTreeRegressor()
```

```python
tree_reg.fit(msft_prepared,msft_labels)
```

```python
msft_dt_predictions = tree_reg.predict(msft_prepared)
tree_mse = mean_squared_error(msft_labels, msft_predictions)
tree_rmse = np.sqrt(tree_mse)

```

```python
tree_rmse
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(tree_reg,msft_prepared,msft_labels, scoring="neg_mean_squared_error", cv=10)

tree_rmse_scores = np.sqrt(-scores)
```

```python
from sklearn.model_selection import GridSearchCV
```
